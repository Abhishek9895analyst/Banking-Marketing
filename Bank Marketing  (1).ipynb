```python
#One Hot Encoding of Categorical Variables
lst = ['job', 'marital','education','default','housing','loan','contact','month','day_of_week','poutcome']
remove = list()
for i in lst:
    if (df[i].dtype == object):
        for j in df[i].unique():
            df[i+'_'+j] = np.where(df[i] == j,1,-1)
        remove.append(i)
df = df.drop(remove, axis=1)
df.head()
```

## Splitting Train and Test Data

```python
# Split Train, test data
df_train = df.sample(frac=0.8,random_state=200)
df_test = df.drop(df_train.index)
```

```python
continuous_vars = ['age', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed', 'duration']
```

```python
# Normailising the Training set
minVec = df_train[continuous_vars].min().copy()
maxVec = df_train[continuous_vars].max().copy()
df_train[continuous_vars] = (df_train[continuous_vars]-minVec)/(maxVec-minVec)
df_train.head()
```

```python
# Test set rescaled using training set limits 
df_test[continuous_vars] = (df_test[continuous_vars]-minVec)/(maxVec-minVec)
df_test.head()
```

```python
# Segregating the attributes and prediction for training and test set
X_train = df_train.loc[:, df_train.columns != 'y']
y_train = df_train.y
X_test = df_test.loc[:, df_test.columns != 'y']
y_test = df_test.y
```

# Decision Tree

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
start = time.process_time()
tracemalloc.start()

dtree = DecisionTreeClassifier(criterion='gini')
dtree.fit(X_train, y_train)
dtreepred = dtree.predict(X_test)

end = time.process_time()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

```

```python
print('Decision Tree')
print('Confusion Matrix\n',confusion_matrix(y_test, dtreepred))
print('Reports\n',classification_report(y_test, dtreepred))
#print('Accuracy\n',round(accuracy_score(y_test, dtreepred),2)*100)
DTREECV = (cross_val_score(dtree, X_train, y_train, cv=k_fold, n_jobs=1, scoring = 'accuracy').mean())
print('Cross Validation Score\n', round(DTREECV, 2))
print('Time taken\n', end - start)
print(f"Current memory usage is {current / 10**3}KB; Peak was {peak / 10**3}KB; Diff = {(peak - current) / 10**3}KB")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
start = time.process_time()
tracemalloc.start()

rfc = RandomForestClassifier(n_estimators = 200)#criterion = entopy,gini
rfc.fit(X_train, y_train)
rfcpred = rfc.predict(X_test)

end = time.process_time()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print('Random Forest')
print('Confusion Matrix\n',confusion_matrix(y_test, rfcpred ))
#print('Accuracy\n',round(accuracy_score(y_test, rfcpred),2)*100)
print('Reports\n',classification_report(y_test, rfcpred))
RFCCV = (cross_val_score(rfc, X_train, y_train, cv=k_fold, n_jobs=1, scoring = 'accuracy').mean())
print('Cross Validation Score\n', round(RFCCV, 2))
print('Time taken\n', end - start)
print(f"Current memory usage is {current / 10**3}KB; Peak was {peak / 10**3}KB; Diff = {(peak - current) / 10**3}KB")
```
